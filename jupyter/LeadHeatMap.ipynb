```python
import logging
import joblib
from sklearn.metrics import log_loss
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd


logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
df = pd.read_csv("../data/datasets/lead_prepared.csv")
```

```python
verbatim_cols = list(filter(lambda x: 'verbatim' in x, df.columns))
status_cols = list(filter(lambda x: 'status' in x, df.columns))
description_cols = list(filter(lambda x: 'description' in x, df.columns))
other_cols = ['year_entered', 'month_entered', 'day_entered', 'year_modified', 'month_modified', 'day_modified']
converted = ['converted']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

plt.subplots(figsize=(16,16))

corr = df[verbatim_cols + converted].corr()
heatmap = sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap="YlGnBu")
heatmap.figure.savefig("lead_corr_matrix_verbatim_converted.png")
```

```python
plt.subplots(figsize=(16,16))

corr = df[description_cols + converted].corr()
heatmap = sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap="YlGnBu")
heatmap.figure.savefig("lead_corr_matrix_description_converted.png")
```

```python
plt.subplots(figsize=(16,16))

corr = df[status_cols + converted].corr()
heatmap = sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap="YlGnBu")
heatmap.figure.savefig("lead_corr_matrix_status_converted.png")
```
