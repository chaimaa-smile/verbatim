```python
from gensim.corpora import WikiCorpus, MmCorpus
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from pprint import pprint
import multiprocessing
```

```python
wiki = WikiCorpus('../data/datasets/frwiki-latest-pages-articles.xml.bz2')
```

```python
class TaggedWikiDocument(object):
    def __init__(self, wiki):
        self.wiki = wiki
        self.wiki.metadata = True
    def __iter__(self):
        for content, (page_id, title) in self.wiki.get_texts():
            yield TaggedDocument([c for c in content], [title])
```

```python
documents = TaggedWikiDocument(wiki)
```

```python
wiki
```

```python
cores = multiprocessing.cpu_count()

model = Doc2Vec(dm=0, dbow_words=1, vector_size=200, window=8, min_count=19, epochs=10, workers=cores)
```

```python
model.build_vocab(documents)
print(str(model))
```

```python
%time model.train(documents, total_examples=model.corpus_count, epochs=10)
model.save("../data/models/wiki_doc2vec200.model")
```
